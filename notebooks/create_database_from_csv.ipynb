```python
import pandas as pd
import sqlite3
```

```python
# Connect to SQLite database (creates it if not exists)
conn = sqlite3.connect("../data/legal_sc.db")

# Load CSV files
c_citation_df = pd.read_csv("../data/case_by_citation/SCDB_2025_01_caseCentered_Citation.csv")
c_docket_df = pd.read_csv("../data/case_by_docket/SCDB_2025_01_caseCentered_Docket.csv")
c_issue_df = pd.read_csv("../data/case_by_issue_regular/SCDB_2025_01_caseCentered_LegalProvision.csv")
c_vote_df = pd.read_csv("../data/case_by_issue_split_vote/SCDB_2025_01_caseCentered_Vote.csv")

j_citation_df = pd.read_csv("../data/justice_by_citation/SCDB_2025_01_justiceCentered_Citation.csv")
j_docket_df = pd.read_csv("../data/justice_by_docket/SCDB_2025_01_justiceCentered_Docket.csv")
j_issue_df = pd.read_csv("../data/justice_by_issue_regular/SCDB_2025_01_justiceCentered_LegalProvision.csv")
j_vote_df = pd.read_csv("../data/justice_by_issue_split_vote/SCDB_2025_01_justiceCentered_Vote.csv")

# Write to SQLite
c_citation_df.to_sql("c_citations", conn, if_exists="replace", index=False)
c_docket_df.to_sql("c_dockets", conn, if_exists="replace", index=False)
c_issue_df.to_sql("c_issues", conn, if_exists="replace", index=False)
c_vote_df.to_sql("c_votes", conn, if_exists="replace", index=False)
j_citation_df.to_sql("j_citations", conn, if_exists="replace", index=False)
j_docket_df.to_sql("j_dockets", conn, if_exists="replace", index=False)
j_issue_df.to_sql("j_issues", conn, if_exists="replace", index=False)
j_vote_df.to_sql("j_votes", conn, if_exists="replace", index=False)

print("CSV files successfully loaded into SQLite")

conn.close()
```
