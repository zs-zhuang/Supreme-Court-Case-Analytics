```python
import requests
import zipfile
import io
import os
```

```python
# data is found at https://scdb.la.psu.edu/data/
# latest data is found at https://scdb.la.psu.edu/data/2025-release-01/
```

```python
def get_sc_data (url, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    # Download ZIP
    response = requests.get(url)
    response.raise_for_status()

    # Open ZIP
    with zipfile.ZipFile(io.BytesIO(response.content)) as z:
        z.extractall(output_dir)

    print("Extracted files:", os.listdir(output_dir))
```

```python
# Download case centered data Organized by Issue/Legal Provision (regular)
url = "https://scdb.la.psu.edu/?jet_download=a3701cc33ed7e29d88d063356a4d98a1be774a09"
output_dir = "../data/case_by_issue_regular"
get_sc_data (url, output_dir)
```
