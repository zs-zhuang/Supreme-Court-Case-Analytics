```python
import pandas as pd
from pathlib import Path
from pprint import pformat
from typing import Any
```

```python
# This function takes a dictionary, write out as a module that can be imported later

def write_mapping_to_module(mapping: dict[Any, Any], variable_name: str, module_path: Path) -> None:
    """
    Writes a Python dictionary to a module file as a variable.

    Args:
        mapping (dict): The dictionary to write.
        variable_name (str): The variable name to assign in the module.
        module_path (Path): Path to the Python file to create/write.
    """
    # Ensure the directory exists
    module_path.parent.mkdir(parents=True, exist_ok=True)

    # Write the dictionary to the Python module
    with module_path.open("w", encoding="utf-8") as f:
        f.write(f"# Auto-generated mapping: {variable_name}\n")
        f.write("# Do not edit by hand\n\n")
        f.write(f"{variable_name} = ")
        f.write(pformat(mapping))
        f.write("\n")
```

```python
url = "http://scdb.wustl.edu/documentation.php?var=justice"

# Read all tables from the page
tables = pd.read_html(url)

# Inspect how many tables were found
print(len(tables))

# Inspect all tables's shape to see which one has the info (looksl ike table 5)
for i, t in enumerate(tables):
    print(f"Table {i} shape:", t.shape)
print()
    
# The Justice ID table is the 5th one on the page
df = tables[5]

# rename the columns
df.columns = ["justice_id", "short_code", "full_text"]

print(df.tail())
```

```python
# map justice ID code to words

df["justice_name"] = (
    df["full_text"]
    .str.extract(r"^(.+?)\s*\(")[0]
)

JUSTICE_NAME_MAP = dict(zip(df["justice_id"], df["justice_name"]))

#print(JUSTICE_NAME_MAP)

# write this dictionary into a module for future use
write_mapping_to_module(JUSTICE_NAME_MAP, "JUSTICE_NAME_MAP", Path("../scdb_code_map/justice_code.py"))
```
